```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mengary=pd.read_csv('train.csv')
```

```python
mengary.head()
```

```python
mengary.describe()
```

```python
loc = pd.get_dummies(mengary['location'],drop_first=True)
cl = pd.get_dummies(mengary['class'],drop_first=True)
seg = pd.get_dummies(mengary['segment'],drop_first=True)
d_type = pd.get_dummies(mengary['delivery type'],drop_first=True)
s_class = pd.get_dummies(mengary['sub-class'],drop_first=True)
```

```python
mengary.drop(['location','class','segment','delivery type','departure city','departure state','placement date','address code','delivery date','sub-class','id'],axis=1,inplace=True)
```

```python
mengary.head()
```

```python
mengary1 = pd.concat([loc,cl,seg,d_type,s_class],axis=1)
```

```python
mengary2=pd.concat([mengary,mengary1],axis=1)
```

```python
mengary2.head()
```

```python
X=mengary2.drop(['profit'],axis=1)
```

```python
Y=mengary2['profit']
```

```python
from sklearn.ensemble import RandomForestRegressor
rfc = RandomForestRegressor(n_estimators=1000)
rfc.fit(X,Y)
```

```python
test=pd.read_csv('test.csv')
```

```python
test.head()
```

```python
loc1 = pd.get_dummies(test['location'],drop_first=True)
cl1 = pd.get_dummies(test['class'],drop_first=True)
seg1 = pd.get_dummies(test['segment'],drop_first=True)
d_type1 = pd.get_dummies(test['delivery type'],drop_first=True)
s_class1 = pd.get_dummies(test['sub-class'],drop_first=True)
```

```python
test.drop(['location','class','segment','delivery type','departure city','departure state','placement date','address code','delivery date','sub-class','id'],axis=1,inplace=True)
```

```python
test1 = pd.concat([loc1,cl1,seg1,d_type1,s_class1],axis=1)
```

```python
test2=pd.concat([test,test1],axis=1)
```

```python
test2.head()
```

```python
X_test=test2
```

```python
predictions = rfc.predict(X_test)
```

```python
df=pd.DataFrame({'profit':predictions})
```
